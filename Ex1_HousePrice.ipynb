## Ex1: Dự đoán giá thuê nhà (House Price)
- Nhà 1 phòng 100đô
- Nhà 2 phòng 150đô
- Nhà 3 phòng 200đô
- ...
- Nhà n phòng là 50 + n*50đô

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

### Định nghĩa và compile model

```
model = tf.keras.Sequential([
          keras.layers.Dense(units=1, input_shape=[1])
])
model.compile(loss='mean_squared_error',
              optimizer='sgd')
```

### Dữ liệu

```
num_room = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype=int)
house_price = np.array([100, 150, 200, 250, 300, 350, 400, 450], dtype=float)

x = num_room
y = house_price
```

### Train model

```
history = model.fit(x, y, epochs=500)
print('Finished training model')
```

### Đồ thị Loss function(epoch)

```
import matplotlib.pyplot as plt
plt.xlabel('Epoch')
plt.ylabel('Loss Magnitude')
plt.plot(history.history['loss'])
```

### Dự đoán giá thuê nhà có 10 phòng
Expect xấp xỉ 550đô

```
print(model.predict([10]))
```
